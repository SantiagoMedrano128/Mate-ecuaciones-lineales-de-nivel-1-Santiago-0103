#### Ecuacion 1
 $x + 8 = 15$
 1. Paso 1: Procedimiento matemáticoPlanteamiento: $x + 8 = 15$
 2. Despeje: El 8 está sumando, pasa al otro lado restando.$x = 15 - 8$
 3. Resultado: $x = 7$

```python
# Despejamos x restando 8 al total
a = int(input("Ingrese el valor primer valor: "))
b = int(input("Ingrese el valor segundo valor: "))
x = a - b
print(f"El valor de x es: {x}")
```

#### Ecuacion 3
$3x + 4x = 21$
* Paso 1: Procedimiento matemático
1. Agrupar términos semejantes: Sumamos las $x$.$(3 + 4)x = 21 \rightarrow 7x =21$
2. Despeje: El 7 multiplica a la $x$, pasa dividiendo.$x = \frac{21}{7}$
3. Resultado: $x = 3$

```python
# Sumamos los coeficientes (3 + 4) y dividimos el resultado
a = int(input("Ingrese el valor primer valor: "))
b = int(input("Ingrese el valor segundo valor: "))
c = int(input("Ingrese el valor tercer valor: "))
x = a / (b + c)
print(f"El valor de x es: {int(x)}")
```

#### Ecuacion 5
 $6a - 2a + a = 20$
 * Paso 1: Procedimiento matemático
 1. Agrupar términos semejantes: (Recuerda que $+ a$ es lo mismo que $+ 1a$).$6a - 2a + 1a = 20 \rightarrow 5a = 20$
 2. Despeje: El 5 pasa dividiendo.$a = \frac{20}{5}$
 3. Resultado: $a = 4$

```python
# Operamos los coeficientes de 'a' y dividimos
a = int(input("Ingrese el valor primer valor: "))
b = int(input("Ingrese el valor segundo valor: "))
c = int(input("Ingrese el valor tercer valor: "))
d = int(input("Ingrese el valor cuarto valor: "))
a = a / (b - c + d)
print(f"El valor de a es: {int(a)}")
```

#### Ecuacion 7
$4x + 3 = 2x + 11$
* Paso 1: Procedimiento matemático
1. Agrupar términos semejantes: Pasamos las $x$ a la izquierda y los números a la derecha.$4x - 2x = 11 - 3$
2. Operar:$2x = 8$
3. Despeje:$x = \frac{8}{2}$
4. Resultado: $x = 4$

```python
# Restamos coeficientes de un lado y constantes del otro
a = int(input("Ingrese el valor primer valor: "))
b = int(input("Ingrese el valor segundo valor: "))
c = int(input("Ingrese el valor tercer valor: "))
d = int(input("Ingrese el valor cuarto valor: "))
x = (d - b) / (a - c)
print(f"El valor de x es: {int(x)}")
```

#### Ecuacion 9
$9m + 4 = 5m + 20$
* Paso 1: Procedimiento matemático
1. Agrupar términos semejantes:$9m - 5m = 20 - 4$
2. Operar:$4m = 16$
3. Despeje:$m = \frac{16}{4}$
4. Resultado: $m = 4$

```python
# m = (constante derecha - constante izquierda) / (coeficiente izq - coeficiente der)
a = int(input("Ingrese el valor primer valor: "))
b = int(input("Ingrese el valor segundo valor: "))
c = int(input("Ingrese el valor tercer valor: "))
d = int(input("Ingrese el valor cuarto valor: "))
m = (d - b) / (a - c)
print(f"El valor de m es: {int(m)}")
```
